# BANK CUSTOMER CHURN ANALYSIS
# ONLY CHURN CUSTOMERS WITH PERCENTAGES

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Style
sns.set(style="whitegrid")
```

# LOAD DATASET

```python
df = pd.read_csv("Churn_Modelling.csv")
```

# DATA CLEANING

```python
# Remove unnecessary columns
df.drop(['RowNumber', 'CustomerId', 'Surname'],
        axis=1,
        inplace=True)
```

# FILTER ONLY CHURN CUSTOMERS

```python
# Only customers who exited
churn_df = df[df['Exited'] == 1]
```

```python
print("This analysis only includes churn customers.")
print("Loyal customers are completely removed.")
print("--------------------------------")
```

# TOTAL CHURN CUSTOMERS

```python
total_churn = len(churn_df)

print("\nTotal Churn Customers:", total_churn)
```

# VISUALIZATION 1
# GENDER ANALYSIS WITH PERCENTAGE

```python
gender_percent = (churn_df['Gender']
                  .value_counts(normalize=True) * 100)

print("\nGender Percentage")
print(gender_percent)

plt.figure(figsize=(7,5))

ax = sns.countplot(x='Gender',
                   data=churn_df)

# Add percentage labels
for p in ax.patches:

    percentage = 100 * p.get_height() / total_churn

    ax.annotate(f'{percentage:.1f}%',
                (p.get_x() + p.get_width()/2,
                 p.get_height()),
                ha='center',
                va='bottom',
                fontsize=11)

plt.title("Churn Customers by Gender (%)")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

# VISUALIZATION 2
# GEOGRAPHY ANALYSIS WITH PERCENTAGE

```python
geo_percent = (churn_df['Geography']
               .value_counts(normalize=True) * 100)

print("\nGeography Percentage")
print(geo_percent)

plt.figure(figsize=(8,5))

ax = sns.countplot(x='Geography',
                   data=churn_df)

for p in ax.patches:

    percentage = 100 * p.get_height() / total_churn

    ax.annotate(f'{percentage:.1f}%',
                (p.get_x() + p.get_width()/2,
                 p.get_height()),
                ha='center',
                va='bottom',
                fontsize=11)

plt.title("Churn Customers by Geography (%)")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

# FINAL COMBINED DASHBOARD WITH PERCENTAGE

```python

# ONLY CHURN CUSTOMERS

fig, axes = plt.subplots(2, 2,
                         figsize=(16,12))

total_churn = len(churn_df)

# 1. GENDER


ax1 = sns.countplot(x='Gender',
                    data=churn_df,
                    ax=axes[0,0])

# Add Percentage Labels
for p in ax1.patches:

    percentage = 100 * p.get_height() / total_churn

    ax1.annotate(f'{percentage:.1f}%',
                 (p.get_x() + p.get_width()/2,
                  p.get_height()),
                 ha='center',
                 va='bottom',
                 fontsize=11)

axes[0,0].set_title("Gender Distribution (%)")

# 2. GEOGRAPHY


ax2 = sns.countplot(x='Geography',
                    data=churn_df,
                    ax=axes[0,1])

for p in ax2.patches:

    percentage = 100 * p.get_height() / total_churn

    ax2.annotate(f'{percentage:.1f}%',
                 (p.get_x() + p.get_width()/2,
                  p.get_height()),
                 ha='center',
                 va='bottom',
                 fontsize=11)

axes[0,1].set_title("Geography Distribution (%)")

# 3. PRODUCTS


ax3 = sns.countplot(x='NumOfProducts',
                    data=churn_df,
                    ax=axes[1,0])

for p in ax3.patches:

    percentage = 100 * p.get_height() / total_churn

    ax3.annotate(f'{percentage:.1f}%',
                 (p.get_x() + p.get_width()/2,
                  p.get_height()),
                 ha='center',
                 va='bottom',
                 fontsize=11)

axes[1,0].set_title("Products Used (%)")

# 4. ACTIVE MEMBERS


ax4 = sns.countplot(x='IsActiveMember',
                    data=churn_df,
                    ax=axes[1,1])

for p in ax4.patches:

    percentage = 100 * p.get_height() / total_churn

    ax4.annotate(f'{percentage:.1f}%',
                 (p.get_x() + p.get_width()/2,
                  p.get_height()),
                 ha='center',
                 va='bottom',
                 fontsize=11)

axes[1,1].set_title("Active Member Status (%)")

# DASHBOARD TITLE


fig.suptitle("Customer Churn Dashboard",
             fontsize=18)

plt.tight_layout()

plt.show()
```

```python
# ORDERED FUNNEL CHART
# ONLY CHURN CUSTOMERS

import plotly.graph_objects as go
# FUNNEL STAGES IN PROPER ORD
total_customers = len(churn_df)

inactive_customers = len(
    churn_df[churn_df['IsActiveMember'] == 0]
)

germany_customers = len(
    churn_df[
        (churn_df['IsActiveMember'] == 0) &
        (churn_df['Geography'] == 'Germany')
    ]
)

one_product_customers = len(
    churn_df[
        (churn_df['IsActiveMember'] == 0) &
        (churn_df['Geography'] == 'Germany') &
        (churn_df['NumOfProducts'] == 1)
    ]
)

female_customers = len(
    churn_df[
        (churn_df['IsActiveMember'] == 0) &
        (churn_df['Geography'] == 'Germany') &
        (churn_df['NumOfProducts'] == 1) &
        (churn_df['Gender'] == 'Female')
    ]
)

# LABELS
funnel_labels = [
    "Total Churn Customers",
    "Inactive Customers",
    "Inactive Germany Customers",
    "Germany Customers with 1 Product",
    "Female Germany Customers with 1 Product"
]

# VALUES

funnel_values = [
    total_customers,
    inactive_customers,
    germany_customers,
    one_product_customers,
    female_customers
]

# CREATE FUNNEL CHART
fig = go.Figure(go.Funnel(

    y = funnel_labels,

    x = funnel_values,

    textinfo = "value+percent initial"

))
# LAYOUT
fig.update_layout(

    title = "Ordered Germany Customer Churn Funnel Analysis",

    font = dict(size = 14)

)
# SHOW CHART
fig.show()
```

```python

# COMPLETE ORDERED FUNNEL CHART
# ONLY CHURN CUSTOMERS

import plotly.graph_objects as go

# TOTAL CHURN CUSTOMERS
total_customers = len(churn_df)

# REGION-WISE CHURN

germany_customers = len(
    churn_df[churn_df['Geography'] == 'Germany']
)

france_customers = len(
    churn_df[churn_df['Geography'] == 'France']
)

spain_customers = len(
    churn_df[churn_df['Geography'] == 'Spain']
)

# GENDER-WISE CHURN
female_customers = len(
    churn_df[churn_df['Gender'] == 'Female']
)

male_customers = len(
    churn_df[churn_df['Gender'] == 'Male']
)

# ACTIVE / INACTIVE CUSTOMERS
inactive_customers = len(
    churn_df[churn_df['IsActiveMember'] == 0]
)

active_customers = len(
    churn_df[churn_df['IsActiveMember'] == 1]
)

# BALANCE ANALYSIS

high_balance_customers = len(
    churn_df[churn_df['Balance'] > 100000]
)

medium_balance_customers = len(
    churn_df[
        (churn_df['Balance'] >= 50000) &
        (churn_df['Balance'] <= 100000)
    ]
)

low_balance_customers = len(
    churn_df[churn_df['Balance'] < 50000]
)

# PRODUCT ANALYSIS

one_product_customers = len(
    churn_df[churn_df['NumOfProducts'] == 1]
)

two_product_customers = len(
    churn_df[churn_df['NumOfProducts'] == 2]
)

# FUNNEL LABELS

funnel_labels = [

    "Total Churn Customers",

    "Germany Customers",
    "France Customers",
    "Spain Customers",

    "Female Customers",
    "Male Customers",

    "Inactive Customers",
    "Active Customers",

    "High Balance Customers",
    "Medium Balance Customers",
    "Low Balance Customers",

    "1 Product Customers",
    "2 Product Customers"
]
# FUNNEL VALUES

funnel_values = [

    total_customers,

    germany_customers,
    france_customers,
    spain_customers,

    female_customers,
    male_customers,

    inactive_customers,
    active_customers,

    high_balance_customers,
    medium_balance_customers,
    low_balance_customers,

    one_product_customers,
    two_product_customers
]

# SORT VALUES IN DESCENDING ORDER

sorted_data = sorted(
    zip(funnel_values, funnel_labels),
    reverse=True
)

sorted_values = [x[0] for x in sorted_data]
sorted_labels = [x[1] for x in sorted_data]

# CREATE FUNNEL CHART


fig = go.Figure(go.Funnel(

    y = sorted_labels,

    x = sorted_values,

    textinfo = "value+percent initial"

))

# UPDATE LAYOUT


fig.update_layout(

    title = "Complete Customer Churn Funnel Analysis",

    font = dict(size = 14),

    height = 700

)

# SHOW FUNNEL CHART


fig.show()
```
